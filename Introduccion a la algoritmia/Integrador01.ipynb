**Ejercicio Stock de Productos**

    Nombre (del producto)

    Categoría (1-Limpieza, 2-Alimentos, 3-Bazar)

    Precio

    Cantidad (en stock)

Se le pide desarrolle un programa que cuente con un menú con las siguientes opciones:

    Indicar la cantidad de productos de la categoría 3 (bazar).

    Indicar el nombre del producto con menor stock en la categoría 2 (alimentos).

    Indicar el nombre del producto más caro de cada una de las categorías

```python

#Librerias

import random as rn

#Funciones

def llenar_vector_random():
    '''Funcion que llena vector de stock y precios con valores random'''
    a = []
    b = []
    for i in range(20):
        numero_random = rn.randint(0,1000)
        a.append(numero_random)
    for p in range(20):
        numero_random = rn.randint(0,100)
        b.append(numero_random)
    return a, b

def cant_bazar(vector_categoria):
    elementos = len(vector_categoria)
    cont = 0
    for i in range(elementos):
        if vector_categoria[i] == 3:
            cont += 1
    print(f'La cantidad de productos de bazar es {cont}')

def Cambio_Precio(vector_precio,vector_nombres):
    a = input(f'Ingresar el nombre del producto: ')
    b = int(input(f'Ingresar el precio a cambiar: '))
    a = a.lower()
    pos = 0
    flag = False
    for i in range(len(vector_nombres)):
        if vector_nombres[i] == a:
            pos = i
            vector_precio.pop(pos)
            vector_precio.insert(pos, b)
            flag = True
    if flag == False:    
        print('El nombre ingresado no existe')
        return ''
    else:
        return vector_precio

def promedio_categoria(vector_categoria,vector_precio):
    a = int(input('Ingrese categoria 1-Limpieza 2-Alimentos 3-Bazar '))
    acumulador = 0
    cont = 0
    elementos = len(vector_categoria)
    for i in range(elementos):
        if vector_categoria[i] == a:
            acumulador += vector_precio[i]
            cont += 1 
    promedio = acumulador / cont
    return promedio

def menos_stock_alimentos(vector_nombres, vector_stock, vector_categoria):
   primer_2 = True
   minimo = 0
   nombre_minimo = ''
   for i in range(len(vector_stock)):
        if vector_categoria[i] == 2 and primer_2:
           primer_2 = False
           minimo = vector_stock[i]
           nombre_minimo = vector_nombres[i]
        elif vector_categoria[i] == 2 and minimo > vector_stock[i]:
           minimo = vector_stock[i]
           nombre_minimo = vector_nombres[i]
   print(f'El producto {nombre_minimo} es el que tiene menor stock, solo quedan {minimo} ')
   return ''

def prod_mas_caro(vector_categoria,vector_precio,vector_nombres):
    max = 0
    pos = 0
    nombre=''
    for i in range(1,4):
        flag = True
        for p in range(len(vector_precio)):
            if vector_categoria[p] == i and flag:
                max = vector_precio[p]
                pos = p
                nombre = vector_nombres[p]
                flag = False
            elif max < vector_precio[p] and vector_categoria[p] == i:
                max = vector_precio[p]
                pos = p
                nombre = vector_nombres[p]
        print(f"En la categoria {i} el producto con el precio mayor es {nombre} y su precio es {max} ")

def menu():
    vector_precio, vector_stock = llenar_vector_random()
    numero = 0
    while numero != 6:
        if numero == 0:
            print('Ingrese 1 para ver la cantidad de productos de la categoría 3.')
            print('Ingrese 2 para actualizar el precio de un producto')
            print('Ingrese 3 para ver precio promedio de los productos de una categorias')
            print('Ingrese 4 para ver el producto con menor stock de alimentos')
            print('Ingrese 5 para ver el producto mas caro de cada categoria')
            print('Ingrese 6 para salir del programa')
            print()
        numero = int(input('Ingrese numero para acceder a la funcion '))        
        if numero == 1:
            cant_bazar(vector_categoria)
        elif numero == 2:
            Cambio_Precio(vector_precio,vector_nombres)
            print(vector_precio)
        elif numero == 3:
            a = promedio_categoria(vector_categoria, vector_precio)
            print(a)
        elif numero == 4:
            print(menos_stock_alimentos(vector_nombres, vector_stock, vector_categoria))
        elif numero == 5:
            prod_mas_caro(vector_categoria,vector_precio,vector_nombres)
        elif numero == 6:
            print('Chau')
            break

#Variable globales
vector_nombres = ['tenedor','platos','tomate','arroz','desinfectante','jabon','cuchillos','avena','fideos','carne','pollo','vasos','escoba','pala','manteles','lechuga','dulce de leche','mate','lavandina','secador de piso' ]
vector_categoria = [3,3,2,2,1,1,3,2,2,2,2,3,3,3,3,2,2,3,1,1]
#Codigo main
menu()



```
