```python
#ejercicio 1 

""" Desarrollar cada una de las siguientes funciones y escribir un programa que permita verificar su funcionamiento, imprimiendo la matriz luego de invocar a cada función: 

f. Calcular el promedio de los elementos de una fila, cuyo número se recibe como 
parámetro. 

g. Calcular el porcentaje de elementos con valor impar en una columna, cuyo nú 
mero se recibe como parámetro. 

h. Determinar si la matriz es simétrica con respecto a su diagonal principal.

i. Determinar si la matriz es simétrica con respecto a su diagonal secundaria. 

j. Determinar qué columnas de la matriz son palíndromos (capicúas), devolviendo una lista con los números de las mismas. 
OG n una NOTA: El valor de N debe leerse por teclado. Las funciones deben servir cualquiera sea el valor ingresado.  """


from random import randint

def matriz_n_x_n(filas,columnas):
    """Cargar números enteros en una matriz de N x N, ingresando los datos desde teclado.  """
    matriz = []
    for i in range(columnas):
        matriz.append([])
        for j in range(filas):
            matriz[i].append(randint(0,9)) 
    #TIENEN QUE INGRESARSE POR TECLADO NO SE OLVIDEN
    return matriz

def ordenar_ascendente(matriz):
    """Ordenar en forma ascendente cada una de las filas de la matriz.""" 
    for i in matriz:
        i.sort()
    return matriz
          
def intercambio_filas(matriz,a,b):
    '''Intercambiar dos filas, cuyos números se reciben como parámetro.
    '''
    aux = matriz[a - 1]
    matriz[a - 1] = matriz[b - 1]
    matriz[b - 1] = aux
    return matriz

def intercambio_columnas(matriz,a,b):
    """Intercambiar dos columnas dadas, cuyos números se reciben como parámetro. """
    for i in matriz: 
        for j in range(len(matriz)):    
            if j == a - 1:  
                aux = i[j] 
                i[j] = i[b - 1]
                i[b - 1] = aux
    return matriz   
        
def transponer_matriz(matriz):
    """e. Trasponer la matriz sobre si misma. 
    (intercambiar cada elemento Aij por Aji) """    
    matriz2 = []
    for i in range(len(matriz[0])):  #len(matriz[0]) = columnas
        matriz2.append([])
    for i in range(0,len(matriz)): #len(matriz) = filas
        for j in range(0,len(matriz[0])): #columnas
            matriz2[i].append(matriz[j][i]) 
    return matriz2
    
    
    
    """
        1 2 3           
    1 [[9 2 5]  1
    2  [3 2 1]  2
    3  [4 5 6]] 3
        1 2 3
        
        1 2 3           
    1 [[9 3 4]  1
    2  [2 2 5]  2
    3  [5 1 6]] 3
        1 2 3
"""
matriz = matriz_n_x_n(3,3)
print(matriz)
assert(ordenar_ascendente(matriz))
assert(intercambio_filas(matriz,1,3))
assert(intercambio_columnas(matriz,1,3))
assert(transponer_matriz(matriz))
```

```python
#Ejercicio 6
def normalizar(lista):
    ''' Escribir una función que reciba una lista de números enteros como parámetro y la 
    normalice, es decir que todos sus elementos deben sumar 1.0, respetando las pro porciones relativas que cada elemento tiene en la lista original.
    Desarrollar tam bién un programa que permita verificar el comportamiento de la función. Por ejem plo, normalizar([1, 1, 2]) debe devolver [0.25, 0.25, 0.50]. 
    '''
    total = 0
    lista2 = []
    for i in lista:
        total += i
    for j in lista:
        resultado = j / total
        lista2.append(resultado)
    return (total, lista2)
lista = [1, 4, 2]
normalizar(lista)
```

```python
#Ejercicio 7       
def intercalar(lista1, lista2):
    '''esta funcion intercala elementos de una lista, Por ejemplo, si lista1 = [8, 1, 3] y lista2 = 
    [5, 9, 7], listal deberÃ¡ quedar como [8, 5, 1, 9, 3, 7]. '''
    for i in range(len(lista2)):
        lista1[(i*2+1):(i*2+1)] = lista2[i:(i+1)]
    return lista1

lista1 = [7,9,1]
lista2 = [3,6,4]
print(intercalar(lista1,lista2))
```
