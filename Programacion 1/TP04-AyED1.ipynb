```python
#Ejercicio 1
def string_capicua(a='AbAAbA'):
    limite_alto = len(a)-1
    limite_bajo = 0
    flag = True
    for i in range(len(a)//2):
        if limite_bajo == limite_alto:
                break
        elif a[limite_bajo] != a[limite_alto]:
            flag = False
            break
        limite_alto -= 1
        limite_bajo += 1          
    return flag
string_capicua()
```

```python
#Ejercicio 5
#Ciclos normales
def filtrar_palabras(cadena,numero):
    lista = cadena.split(' ')
    lista2= []
    for i in lista:
        if len(i) >= numero:
            lista2.append(i)
    lista2 = ' '.join(lista2)
    return lista2

#Lista por comprension
def filtrar_palabras2(cadena,numero):
    lista = cadena.split(' ')
    lista2= [i for i in lista if len(i)>=numero]
    return lista2

#Por filter
def filtrar_palabras3(cadena, numero):
    lista = cadena.split(' ')
    l = list(filter(lambda z: len(z) >= numero ,lista))
    return l
filtrar_palabras3('La vida es bella', 4)
```

```python
#Ejercicio 9
def ultimos_caracteres(cadena,numero):
    numero = len(cadena) - numero
    contador = 0
    lista = []
    for i in cadena:
        contador += 1
        if contador <= numero:
            lista.append(i)
    lista = ''.join(lista)
    return lista
ultimos_caracteres('la vida es bella', 4)
```

```python
#Ejercicio 10
def reemplazo_palabras(cadena, palabra,reemplazo):
    lista = cadena.split(' ')
    lista2 = []
    contador_reemplazo = 0
    for i in lista:
        if i == palabra:
            lista2.append(reemplazo)
            contador_reemplazo += 1
        else: 
            lista2.append(i)
    lista2 = ' '.join(lista2)
    return lista2, contador_reemplazo
reemplazo_palabras('La vida es bella bella', 'bella','fea')
```
